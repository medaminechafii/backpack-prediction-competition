```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load the data
mnist  = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train,x_test = X_train/255.0, X_test/255.0
```

```python
print(X_train.shape)
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),
                                   tf.keras.layers.Dense(120,activation = "relu"),
                                   tf.keras.layers.Dropout(0.2),
                                   tf.keras.layers.Dense(10,activation = "softmax")])
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
r = model.fit(X_train,y_train,validation_data = (X_test,y_test), epochs = 10)
```

```python
plt.plot(r.history["loss"], label = "loss")
plt.plot(r.history["val_loss"], label = "val_loss")
plt.legend()
```

```python
plt.plot(r.history["accuracy"], label = "accuracy")
plt.plot(r.history["val_accuracy"], label = "val_accuracy")
plt.legend()
```

```python
print(model.evaluate(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
import itertools
def plot_confusion_matrix(cv,classes,normalize = False,title = "Confusion Matrix",cmap = plt.cm.Blues):
    if normalize:
        cv = cv.astype("float")/cv.sum(axis = 1)[:,np.newaxis]
    else:
        print("Confusion Matrix without normalization")
    print(cv)
    plt.imshow(cv,interpolation = "nearest",cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation = 45)
    plt.yticks(tick_marks,classes)
    fmt = ".2f" if normalize else "d"
    thresh = cv.max()/2
    for i,j in itertools.product(range(cv.shape[0]),range(cv.shape[1])):
        plt.text(j,i,format(cv[i,j],fmt),horizontalalignment ="center",color = "white" if cv[i,j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.show()
p_test = model.predict(X_test).argmax(axis = 1)
cm = confusion_matrix(y_test,p_test)
plot_confusion_matrix(cm,list(range(10)))
```

```python
miscalculated = np.where(p_test != y_test)[0]
i = np.random.choice(miscalculated)
plt.imshow(X_test[i],cmap = "gray")
plt.title("Predicted Label: %d"%p_test[i])
plt.show()
```
